```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the pre-trained model and tokenizer (using GPT-2 in this case)
model_name = 'gpt2'
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name, clean_up_tokenization_spaces=False)


def calculate_cross_entropy(sentence):
    # Encode the sentence
    tokens = tokenizer.encode(sentence, return_tensors='pt')

    # Get the model's predictions (logits) for the sentence
    with torch.no_grad():
        outputs = model(tokens, labels=tokens)

    # The loss is the cross-entropy of the sentence
    # outputs.loss returns the average loss, but we need to compute it step by step
    loss = outputs.loss.item()

    # Return the cross-entropy H(x)
    return loss


def fluency_score(sentence):
    # Compute H(x)
    Hx = calculate_cross_entropy(sentence)
    print(Hx)

    # Compute f(x)
    fx = 1 / (1 + Hx)

    return fx


# Example sentence
orig_sentence = "From this scope , social media has shorten our distance ."
hypo_sentence = "From this scope , social media has shortened our distance ."
gold_sentence = "From this perspective , social media has shortened the distance between us ."

# Compute the fluency score
ori_score = fluency_score(orig_sentence)
hyp_score = fluency_score(hypo_sentence)
ref_score = fluency_score(gold_sentence)

print(f"Fluency score for the original sentence: {ori_score:.4f}")
print(f"Fluency score for the hypothesis sentence: {hyp_score:.4f}")
print(f"Fluency score for the reference sentence: {ref_score:.4f}")
```

```python
import os

def fluency_score(sentence):
    # Compute H(x)
    Hx = calculate_cross_entropy(sentence)

    # Compute f(x)
    fx = 1 / (1 + Hx)

    return fx

# Function to process all text files in the directory and compute fluency scores
def compute_fluency_for_systems(directory, ref_file_0, ref_file_1):
    system_fluency = {}

    # Loop through all the files in the directory
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            system_name = filename.split('.')[0]  # Extract system name from the filename
            print('-' * 50)
            print(f"Processing system: {system_name}")
            file_path = os.path.join(directory, filename)

            # Read the file and compute fluency for each sentence
            with open(file_path, 'r') as file, open(ref_file_1, 'r') as ref_1, open(ref_file_0, 'r') as ref_0:
                sentences = file.readlines()
                ref_0 = ref_0.readlines()
                ref_1 = ref_1.readlines()

            fluency_values = []
            ref_0_fluency = []
            ref_1_fluency = []
            
            for sentence, ref_0_sent, ref_1_sent in zip(sentences, ref_0, ref_1):
                sentence = sentence.strip()  # Remove leading/trailing whitespace
                ref_0_sent = ref_0_sent.strip()
                ref_1_sent = ref_1_sent.strip()
                
                if sentence:
                    fluency_values.append(fluency_score(sentence) / (fluency_score(ref_0_sent) + fluency_score(ref_1_sent)))

            # Compute average fluency score for the system
            if fluency_values:
                system_fluency[system_name] = sum(fluency_values) / len(fluency_values)

            # Export sentence-level fluency scores to a separate file
            export_sentence_fluency(system_name, fluency_values)

    return system_fluency

# Function to export the fluency values of each sentence into a separate file
def export_sentence_fluency(system_name, fluency_values):
    # Create the directory if it doesn't exist
    sentence_fluency_dir = "sentence_relative_fluency"
    if not os.path.exists(sentence_fluency_dir):
        os.makedirs(sentence_fluency_dir)

    # File path for the system's sentence-level fluency scores
    output_file_path = os.path.join(sentence_fluency_dir, f"{system_name}.txt")

    # Write the fluency scores to the system's file
    with open(output_file_path, 'w') as f:
        for score in fluency_values:
            f.write(f"{score:.4f}\n")

# Function to export the fluency scores to a text file
def export_fluency_scores(system_fluency, output_file):
    # Sort the systems alphabetically by their name
    sorted_systems = sorted(system_fluency.keys())

    # Write the fluency scores to the output file
    with open(output_file, 'w') as f:
        for system in sorted_systems:
            f.write(f"{system_fluency[system]:.4f}\n")

# Example usage
directory = "subset"  # The directory containing the .txt files for each system
output_file = "subset_relative_fluency.txt"  # The output file to save the fluency scores

system_fluency = compute_fluency_for_systems(directory, 'golden_correction_0_annotator.txt', 'golden_correction_1_annotator.txt')
export_fluency_scores(system_fluency, output_file)

print(f"Fluency scores have been computed and saved to {output_file}")
```

```python
import os
import numpy as np

# Define the directory where the .txt files are stored
directory = "GJG15_sentence_fluency"

# Function to read the file and compute the average of valid scores
def compute_average(file_path):
    scores = []
    with open(file_path, 'r') as file:
        for line in file:
            try:
                score = float(line.strip())
                if not np.isnan(score):  # Ignore 'nan'
                    scores.append(score*4)
            except ValueError:
                continue  # Skip non-numeric lines
    if scores:
        return np.mean(scores)
    return None  # If no valid scores were found

# Function to process all .txt files in the directory
def process_files(directory):
    averages = []
    filenames = []

    for filename in sorted(os.listdir(directory)):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)
            average = compute_average(file_path)
            if average is not None:
                averages.append(average)
                filenames.append(filename)

    return zip(filenames, averages)

# Write the averages to a new file
def write_averages_to_file(output_file, directory):
    with open(output_file, 'w') as f:
        for filename, average in process_files(directory):
            f.write(f"{average}\n")

# Set the output file path
output_file = "GJG15_fluency_scores.txt"

# Call the function to write the averages
write_averages_to_file(output_file, directory)

print(f"Averages written to {output_file}")
```

```python
import matplotlib.pyplot as plt
import math
import numpy as np
import seaborn as sns

# Set the style for a professional look
plt.style.use('seaborn-v0_8-paper')
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.weight'] = 'bold'
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['axes.titleweight'] = 'bold'

def plot_score_distribution(file_path, bins=50):
    # Read scores from file
    with open(file_path, 'r') as file:
        scores = [float(line.strip()) for line in file]
    
    # Create figure and axis with professional dimensions
    plt.figure(figsize=(10, 6), dpi=300)
    
    # Create histogram
    n, bins, patches = plt.hist(scores, bins=bins, density=True, 
                              color='#4C72B0', alpha=0.75,
                              edgecolor='black', linewidth=1)
    
    # Add kernel density estimation
    sns.kdeplot(data=scores, color='#C44E52', linewidth=2,
                label='Density Estimation')
    
    # Add mean and median lines
    # mean_score = np.mean(scores)
    # median_score = np.median(scores)

    scores = [0 if np.isnan(x) else x for x in scores]
    scores = [math.exp(score) for score in scores]
    print(scores)
    
    mean_score = sum(scores) / len(scores)
    median_score = np.median(scores)
    
    plt.axvline(mean_score, color='#55A868', linestyle='--', linewidth=2,
                label=f'Mean: {mean_score:.2f}')
    plt.axvline(median_score, color='#8172B3', linestyle=':', linewidth=2,
                label=f'Median: {median_score:.2f}')
    
    # Customize plot
    plt.title('Distribution of Scores', pad=15, fontsize=14)
    plt.xlabel('Score', labelpad=10, fontsize=12)
    plt.ylabel('Density', labelpad=10, fontsize=12)
    plt.legend(frameon=True, facecolor='white', edgecolor='none',
              fontsize=10, loc='upper right')
    
    # Add grid for better readability
    plt.grid(True, linestyle='--', alpha=0.3)
    
    # Adjust layout
    plt.tight_layout()
    
    # Add statistical information
    stats_text = f'n = {len(scores)}\n'
    stats_text += f'μ = {mean_score:.2f}\n'
    stats_text += f'σ = {np.std(scores):.2f}'
    plt.text(0.95, 0.95, stats_text,
            transform=plt.gca().transAxes,
            verticalalignment='top',
            horizontalalignment='left',
            bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'),
            fontsize=10)
    
    # Save the plot with high resolution
    # plt.savefig('score_distribution.png', dpi=300, bbox_inches='tight')
    # plt.close()
    plt.show()

# Example usage
if __name__ == "__main__":
    plot_score_distribution('reference_fluency/ref0.txt')
```

```python
'''
Here begins new evaluation method trials for fluency
'''

import os
import numpy as np
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from tqdm import tqdm

# ---------------------------
# Setup: load pretrained model and tokenizer
# ---------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_name = "gpt2"  # using GPT-2 as the mainstream pretrained model
model = GPT2LMHeadModel.from_pretrained(model_name)
model.to(device)
model.eval()  # set model to evaluation mode

tokenizer = GPT2Tokenizer.from_pretrained(model_name)
# GPT-2 does not have a pad_token by default; set it to the EOS token if missing.
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# ---------------------------
# Function to compute cross-entropy (Hx) for a given sentence.
# Here, we use the model’s loss (average negative log probability per token).
# ---------------------------
def calculate_cross_entropy(sentence: str) -> float:
    """Compute cross-entropy for a sentence using GPT-2.
    
    Args:
        sentence (str): The input sentence.
        
    Returns:
        float: The average negative log likelihood (cross-entropy) per token.
    """
    # Skip empty sentences.
    if not sentence.strip():
        return None
    
    # Tokenize the sentence.
    inputs = tokenizer(sentence, return_tensors="pt")
    input_ids = inputs.input_ids.to(device)
    
    # Compute loss using the model (providing labels returns the loss).
    with torch.no_grad():
        outputs = model(input_ids, labels=input_ids)
        loss = outputs.loss  # loss is averaged over the tokens
    
    return loss.item()

# ---------------------------
# Function to compute length-normalized statistics.
# It groups sentences by their token count and computes:
#   - Count of sentences in the group.
#   - Average cross-entropy.
#   - Standard deviation of cross-entropy.
# ---------------------------
def generate_length_stats(sentences: list, scores: list) -> dict:
    """Group sentences by token count and compute statistics.
    
    Args:
        sentences (list): List of sentence strings.
        scores (list): List of corresponding average Hx scores.
    
    Returns:
        dict: Mapping from token count to a dictionary with count, avg, and std of Hx.
    """
    length_dict = {}
    for sent, score in zip(sentences, scores):
        # Tokenize to get token count
        tokens = tokenizer.tokenize(sent)
        token_count = len(tokens)
        if token_count not in length_dict:
            length_dict[token_count] = []
        length_dict[token_count].append(score)
    
    # Compute average and standard deviation for each token length
    length_stats = {}
    for token_count, score_list in length_dict.items():
        length_stats[token_count] = {
            "count": len(score_list),
            "avg_score": np.mean(score_list),
            "std_score": np.std(score_list)
        }
    return length_stats

def standardized_fluency_score(sentence, ref_mean, ref_std, k=2):
    Hx = calculate_cross_entropy(sentence)
    if type(Hx) == float:
        z = (ref_mean - Hx) / ref_std  # Higher z = better fluency
    else:
        z = 0
    return 1 / (1 + np.exp(-k * z))  # Sigmoid scaling

# ---------------------------
# Main processing function.
# ---------------------------
def main():
    # Directory that contains the reference text files.
    system_output_dire = "Subset/subset"
    ref_dir = "Subset/reference"
    
    # List all .txt files in the directory.
    ref_files = [os.path.join(ref_dir, f) for f in os.listdir(ref_dir) if f.endswith(".txt")]
    ref_files = sorted(ref_files)  # sort for consistent order
    
    # Ensure that at least two reference files are found.
    if len(ref_files) < 2:
        print("Error: need at least 2 reference .txt files in the 'reference' directory.")
        return
    
    # Read lines from the first two reference files.
    with open(ref_files[0], "r", encoding="utf-8") as f:
        lines1 = f.readlines()
    with open(ref_files[1], "r", encoding="utf-8") as f:
        lines2 = f.readlines()
    
    # Use the minimum number of lines (if the files differ in length).
    num_sentences = min(len(lines1), len(lines2))
    avg_hx_scores = []  # store the average cross entropy for each sentence pair
    all_sentences = []  # store one representative sentence (from the first file) for length stats
    
    # print("Computing cross-entropy for reference sentences...")
    # # Loop over each pair of sentences from the two files.
    # for i in tqdm(range(num_sentences)):
    #     sentence1 = lines1[i].strip()
    #     sentence2 = lines2[i].strip()
        
    #     # Skip if any sentence is empty.
    #     if not sentence1 or not sentence2:
    #         continue
        
    #     # Compute cross-entropy for each sentence.
    #     hx1 = calculate_cross_entropy(sentence1)
    #     hx2 = calculate_cross_entropy(sentence2)
        
    #     # If either computation fails, skip this pair.
    #     if hx1 is None or hx2 is None:
    #         continue
        
    #     # Average the Hx scores from the two files.
    #     avg_hx = (hx1 + hx2) / 2.0
    #     avg_hx_scores.append(avg_hx)
    #     all_sentences.append(sentence1)  # you can choose either sentence as representative
    
    # # Write the average Hx scores to a new file, one score per line.
    # with open("reference_Hx_scores.txt", "w", encoding="utf-8") as f:
    #     for score in avg_hx_scores:
    #         f.write(f"{score}\n") 
    
    # # Compute overall statistics: mean (μ) and standard deviation (σ)
    # avg_hx_scores = [0 if np.isnan(x) else x for x in avg_hx_scores]
    # ref_mean = np.mean(avg_hx_scores)
    # ref_std = np.std(avg_hx_scores)
    # print(f"\nReference Cross-Entropy Mean (μ): {ref_mean:.4f}")
    # print(f"Reference Cross-Entropy Std (σ): {ref_std:.4f}")

    system_files = [os.path.join(system_output_dire, f) for f in os.listdir(system_output_dire) if f.endswith(".txt")]
    for files in system_files:
        with open(files, "r", encoding="utf-8") as f:
            lines = f.readlines()
            for i in tqdm(range(num_sentences)):
                sentence = lines[i].strip()
                standardized_fluency = standardized_fluency_score(sentence, 4.3128, 0.7004)
                filename = os.path.join("Subset/standardized_fluency_scores", os.path.basename(files))
                with open(filename, "a", encoding="utf-8") as f:
                    f.write(f"{standardized_fluency}\n")
    
    # Compute length-normalized statistics.
    length_stats = generate_length_stats(all_sentences, avg_hx_scores)
    
    print("\nLength-normalized statistics (Token Count: Count, Avg Hx, Std Hx):")
    for token_count in sorted(length_stats.keys()):
        stats = length_stats[token_count]
        print(f"Token Count {token_count}: Count = {stats['count']}, "
              f"Avg Hx = {stats['avg_score']:.4f}, Std = {stats['std_score']:.4f}")
    
    # Optionally, write the length-normalized statistics to a file.
    with open("length_normalized_stats.txt", "w", encoding="utf-8") as f:
        f.write("TokenCount\tCount\tAvgHx\tStdHx\n")
        for token_count in sorted(length_stats.keys()):
            stats = length_stats[token_count]
            f.write(f"{token_count}\t{stats['count']}\t{stats['avg_score']:.4f}\t{stats['std_score']:.4f}\n")

# ---------------------------
# Run the main function.
# ---------------------------
if __name__ == "__main__":
    main()
```

```python
'''
Here begins new evaluation method trials for fluency
'''

import os
import numpy as np
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from tqdm import tqdm

# ---------------------------
# Setup: load pretrained model and tokenizer
# ---------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_name = "gpt2"  # using GPT-2 as the mainstream pretrained model
model = GPT2LMHeadModel.from_pretrained(model_name)
model.to(device)
model.eval()  # set model to evaluation mode

tokenizer = GPT2Tokenizer.from_pretrained(model_name)
# GPT-2 does not have a pad_token by default; set it to the EOS token if missing.
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# ---------------------------
# Function to compute cross-entropy (Hx) for a given sentence.
# Here, we use the model’s loss (average negative log probability per token).
# ---------------------------
def calculate_cross_entropy(sentence: str) -> float:
    """Compute cross-entropy for a sentence using GPT-2.
    
    Args:
        sentence (str): The input sentence.
        
    Returns:
        float: The average negative log likelihood (cross-entropy) per token.
    """
    # Skip empty sentences.
    if not sentence.strip():
        return None
    
    # Tokenize the sentence.
    inputs = tokenizer(sentence, return_tensors="pt")
    input_ids = inputs.input_ids.to(device)
    
    # Compute loss using the model (providing labels returns the loss).
    with torch.no_grad():
        outputs = model(input_ids, labels=input_ids)
        loss = outputs.loss  # loss is averaged over the tokens
    
    return loss.item()

# ---------------------------
# Function to compute length-normalized statistics.
# It groups sentences by their token count and computes:
#   - Count of sentences in the group.
#   - Average cross-entropy.
#   - Standard deviation of cross-entropy.
# ---------------------------
def generate_length_stats(sentences: list, scores: list) -> dict:
    """Group sentences by token count and compute statistics.
    
    Args:
        sentences (list): List of sentence strings.
        scores (list): List of corresponding average Hx scores.
    
    Returns:
        dict: Mapping from token count to a dictionary with count, avg, and std of Hx.
    """
    length_dict = {}
    for sent, score in zip(sentences, scores):
        # Tokenize to get token count
        tokens = tokenizer.tokenize(sent)
        token_count = len(tokens)
        if token_count not in length_dict:
            length_dict[token_count] = []
        length_dict[token_count].append(score)
    
    # Compute average and standard deviation for each token length
    length_stats = {}
    for token_count, score_list in length_dict.items():
        length_stats[token_count] = {
            "count": len(score_list),
            "avg_score": np.mean(score_list),
            "std_score": np.std(score_list)
        }
    return length_stats

def standardized_fluency_score(sentence, ref_mean, ref_std, k=2):
    Hx = calculate_cross_entropy(sentence)
    if np.isnan(Hx)
    z = (ref_mean - Hx) / ref_std  # Higher z = better fluency
    return 1 / (1 + np.exp(-k * z))  # Sigmoid scaling

# ---------------------------
# Main processing function.
# ---------------------------
def main():
    # Directory that contains the reference text files.
    system_output_dire = "Subset/official_submission_TXT"
    ref_dir = "Subset/reference"
    
    # List all .txt files in the directory.
    ref_files = [os.path.join(ref_dir, f) for f in os.listdir(ref_dir) if f.endswith(".txt")]
    ref_files = sorted(ref_files)  # sort for consistent order
    
    # Ensure that at least two reference files are found.
    if len(ref_files) < 2:
        print("Error: need at least 2 reference .txt files in the 'reference' directory.")
        return
    
    # Read lines from the first two reference files.
    with open(ref_files[0], "r", encoding="utf-8") as f:
        lines1 = f.readlines()
    with open(ref_files[1], "r", encoding="utf-8") as f:
        lines2 = f.readlines()
    
    # Use the minimum number of lines (if the files differ in length).
    num_sentences = min(len(lines1), len(lines2))
    avg_hx_scores = []  # store the average cross entropy for each sentence pair
    all_sentences = []  # store one representative sentence (from the first file) for length stats
    
    # print("Computing cross-entropy for reference sentences...")
    # # Loop over each pair of sentences from the two files.
    # for i in tqdm(range(num_sentences)):
    #     sentence1 = lines1[i].strip()
    #     sentence2 = lines2[i].strip()
        
    #     # Skip if any sentence is empty.
    #     if not sentence1 or not sentence2:
    #         continue
        
    #     # Compute cross-entropy for each sentence.
    #     hx1 = calculate_cross_entropy(sentence1)
    #     hx2 = calculate_cross_entropy(sentence2)
        
    #     # If either computation fails, skip this pair.
    #     if hx1 is None or hx2 is None:
    #         continue
        
    #     # Average the Hx scores from the two files.
    #     avg_hx = (hx1 + hx2) / 2.0
    #     avg_hx_scores.append(avg_hx)
    #     all_sentences.append(sentence1)  # you can choose either sentence as representative
    
    # # Write the average Hx scores to a new file, one score per line.
    # with open("reference_Hx_scores.txt", "w", encoding="utf-8") as f:
    #     for score in avg_hx_scores:
    #         f.write(f"{score}\n") 
    
    # # Compute overall statistics: mean (μ) and standard deviation (σ)
    # avg_hx_scores = [0 if np.isnan(x) else x for x in avg_hx_scores]
    # ref_mean = np.mean(avg_hx_scores)
    # ref_std = np.std(avg_hx_scores)
    # print(f"\nReference Cross-Entropy Mean (μ): {ref_mean:.4f}")
    # print(f"Reference Cross-Entropy Std (σ): {ref_std:.4f}")

    system_files = [os.path.join(system_output_dire, f) for f in os.listdir(system_output_dire) if f.endswith(".txt")]
    for files in system_files:
        with open(files, "r", encoding="utf-8") as f:
            lines = f.readlines()
            for i in tqdm(range(num_sentences)):
                sentence = lines[i].strip()
                standardized_fluency = standardized_fluency_score(sentence, 4.3128, 0.7004)
                with open(f"{files}.fluency", "a", encoding="utf-8") as f:
                    f.write(f"{standardized_fluency}\n")
    
    # Compute length-normalized statistics.
    length_stats = generate_length_stats(all_sentences, avg_hx_scores)
    
    print("\nLength-normalized statistics (Token Count: Count, Avg Hx, Std Hx):")
    for token_count in sorted(length_stats.keys()):
        stats = length_stats[token_count]
        print(f"Token Count {token_count}: Count = {stats['count']}, "
              f"Avg Hx = {stats['avg_score']:.4f}, Std = {stats['std_score']:.4f}")
    
    # Optionally, write the length-normalized statistics to a file.
    with open("length_normalized_stats.txt", "w", encoding="utf-8") as f:
        f.write("TokenCount\tCount\tAvgHx\tStdHx\n")
        for token_count in sorted(length_stats.keys()):
            stats = length_stats[token_count]
            f.write(f"{token_count}\t{stats['count']}\t{stats['avg_score']:.4f}\t{stats['std_score']:.4f}\n")

# ---------------------------
# Run the main function.
# ---------------------------
if __name__ == "__main__":
    main()
```
